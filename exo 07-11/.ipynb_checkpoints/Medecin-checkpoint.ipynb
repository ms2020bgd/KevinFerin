```python
populationDep = pd.read_excel("estim-pop-dep-sexe-gca-1975-2019.xls",3)
populationDep.columns= ["num_dep",populationDep.iloc[2][0]]+list(populationDep.iloc[3][2:].values)
populationDep = populationDep.drop([0,1,2,3])
populationDep=populationDep.dropna()
```

```python
honoraireMedecin =  pd.read_excel("Honoraires_des_professionnels_de_sante_APE_par_departement_en_2017.xls",3)
honoraireMedecin["dpt_num"]= honoraireMedecin["DEPARTEMENT"].str.split("-").str[0]
honoraireMedecin["dpt"]= honoraireMedecin["DEPARTEMENT"].str.split("-").str[1]
honoraireMedecin["spec_num"]= honoraireMedecin["Généralistes et compétences MEP"].str.split("-").str[0]
honoraireMedecin["spec"]= honoraireMedecin["Généralistes et compétences MEP"].str.split("-").str[1]
honoraireMedecin = honoraireMedecin[['spec','spec_num','dpt','dpt_num',"NOMBRE D'APE",'DEPASSEMENTS (Euros)']]
honoraireMedecin = honoraireMedecin.iloc[:102]
honoraireMedecin=honoraireMedecin.dropna()
l = populationDep[["num_dep","Total"]]
l.columns=["num_dep","Total", "Homme", "Femme"]
l.reset_index()
honoraireMedecin["populationTot"] = l["Total"].values
honoraireMedecin["populationHomme"] = l["Homme"].values
honoraireMedecin["populationFemme"] = l["Femme"].values
```

```python


plt.scatter(x=honoraireMedecin["populationTot"], y =honoraireMedecin["DEPASSEMENTS (Euros)"])
plt.xlabel("Population")
plt.ylabel("Dépassement honoraire")
plt.title("Lien entre population et dépassement honoraires des médecins généralistes par départemtn")
```

On peut clairement voir une corrélation, plus la population
